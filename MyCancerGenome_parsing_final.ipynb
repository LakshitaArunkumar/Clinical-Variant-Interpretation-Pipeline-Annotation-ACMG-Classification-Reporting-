```python
import requests as r
from bs4 import BeautifulSoup
import time
import pandas as pd
from IPython.display import FileLink
```

```python
def For_Genes(name):
    if '-' in name:
        name1 = name[:name.index('-')]
        response = r.get('https://www.mycancergenome.org/content/gene/'+name1+'/')
    else:
        response = r.get('https://www.mycancergenome.org/content/gene/'+name+'/', verify=False)
    r1f = BeautifulSoup(response.content, "html.parser")
    gene_role_paragraphs = r1f.find_all("p")[:3]
    gene_role_text = "".join(paragraph.get_text(strip=True) for paragraph in gene_role_paragraphs)
    body = r1f.find("div", class_="context-container")

    df1 = pd.DataFrame(columns=['Gene_Name', 'Gene or Alteration match', 'Drugs', 'Cancer', 'Biomarker_Criteria', 'Gene_Role'])

    if body is None:
        print('No Gene Match')

    else:
        drugs = body.find_all("div", class_="about-gene-therapy-row")
        gene_role_paragraphs = r1f.find_all("p")[:2]
        gene_role_text = "".join(paragraph.get_text(strip=True) for paragraph in gene_role_paragraphs)
        data = []
        for i in drugs:
            a = i.find("p", class_="about-gene-therapy-header").text.replace('+\n', '').strip()
            for j in i.find_all("div", class_='about-gene-therapy-disease-row'):
                b = j.find("p", class_="about-gene-therapy-disease-header").text.replace('-\n', '').strip()
                c = j.find("div", class_="row table-row targeted-therapy-table-small-screen-container").text.strip().replace('\n\n\n\n', '\n').replace('\n\n\n', '\n')

                data.append([name.upper(), 'Gene match', a, b, c, gene_role_text])
        
        df1 = pd.concat([df1, pd.DataFrame(data, columns=['Gene_Name', 'Gene or Alteration match', 'Drugs', 'Cancer', 'Biomarker_Criteria', 'Gene_Role'])])

    df1 = pd.concat([df1, pd.DataFrame({'Gene_Name': name.upper(), 'Gene or Alteration match': '', 'Drugs': '', 'Cancer': '',
                                        'Biomarker_Criteria': '', 'Gene_Role': gene_role_text}, index=[0])], ignore_index=True)
    
    return df1
```

```python
def For_Alterations(name):
    url = 'https://www.mycancergenome.org/content/alteration/'+name+'/'
    #print("Link:", url)
    time.sleep(2)
    response = r.get(url)
    r1f = BeautifulSoup(response.content,"html.parser")
    #print(r1f)
    drugs = r1f.find_all("div", class_="about-gene-therapy-row")
    #print(drugs)
    gene_role_paragraphs = r1f.find_all("p")[:2]
    #print(gene_role_paragraphs)
    gene_role_text = "".join(paragraph.get_text(strip=True) for paragraph in gene_role_paragraphs)
    df1 = pd.DataFrame(columns=['Gene_Name', 'Gene or Alteration match', 'Drugs', 'Cancer', 'Biomarker_Criteria', 'Gene_Role'])
    data = []
    for i in drugs: 
        a=i.find("p", class_="about-alteration-therapy-header").text.replace('+\n', '').strip()
        for j in i.find_all("div", class_='about-alteration-therapy-disease-row'):
            b=j.find('p', class_="about-alteration-therapy-disease-header").text.replace('-\n', '').strip()
            c=j.find("div", class_="row table-row targeted-therapy-table-small-screen-container").text.strip().replace('\n\n\n\n','\n').replace('\n\n\n','\n')
            data.append([name.upper(), 'Gene match', a, b, c, gene_role_text])
        
        df1 = pd.concat([df1, pd.DataFrame(data, columns=['Gene_Name', 'Gene or Alteration match', 'Drugs', 'Cancer', 'Biomarker_Criteria', 'Gene_Role'])])

    df1 = pd.concat([df1, pd.DataFrame({'Gene_Name': name.upper(), 'Gene or Alteration match': '', 'Drugs': '', 'Cancer': '',
                                        'Biomarker_Criteria': '', 'Gene_Role': gene_role_text}, index=[0])], ignore_index=True)
    
    return df1
```
