```python
def getSampleInformationFromRedmine(redmine_id):
    patient_dict = {}
    issue = issue = redmine.issue.get(redmine_id)
    for custom in issue.custom_fields:
        #print(custom)
        if (custom.name == 'Patient Name'):
            patient_dict['patient_name'] = custom.value
        elif (custom.name == 'Patient Gender'):   
            patient_dict['patient_gender'] = custom.value
        elif (custom.name == 'Patient Age'):
            patient_dict['patient_age'] = custom.value
        elif (custom.name == 'Lilac Identifier'):
            patient_dict['lilac_id'] = custom.value
        elif (custom.name == 'TMB Status'):
            patient_dict['TMB_status'] = custom.value
        elif (custom.name == 'MSI Status'):
            patient_dict['MSI_status'] = custom.value
        elif (custom.name == 'MMR Status'):
            patient_dict['MMR_status'] = custom.value
        elif (custom.name == 'HRR Status'):
            patient_dict['HRR_status'] = custom.value
        elif (custom.name == 'PD-L1 Status'):
            patient_dict['PD-L1_status'] = custom.value
        elif (custom.name == 'PD-L1 IHC status and score'):
            patient_dict['PD-L1_status_score'] = custom.value
        elif (custom.name == 'Conditions for Clinical Trials'):
            patient_dict['ct_conditions'] = custom.value
        elif (custom.name == 'Main Tumor Type'):
            patient_dict['Main Tumor Type'] = custom.value
        elif (custom.name == 'Genes for Clinical Trials'):
            patient_dict['ct_genes'] = custom.value
        elif (custom.name == 'OtherTerms for Clinical Trials'):
            patient_dict['ct_other_terms'] = custom.value
        elif (custom.name == 'FFPE Block Number'):
            patient_dict['FFPE_block_number'] = custom.value
        elif (custom.name == 'Referring Centre'):
            patient_dict['referring_centre'] = custom.value
        elif (custom.name == 'Referring Physician'):
            patient_dict['referring_physician'] = custom.value
        elif (custom.name == 'Sample Collection Date'):
            patient_dict['sample_collection_date'] = custom.value
        elif (custom.name == '% of Tumor Cells'):
            patient_dict['tumor_cells'] = custom.value
        elif (custom.name == 'Metastatic Status'):
            patient_dict['metastatic_status'] = custom.value
        elif (custom.name == 'Metastatic Site'):
            patient_dict['metastatic_site'] = custom.value
        elif (custom.name == 'Histopathology Summary'):
            patient_dict['histopathology_summary'] = custom.value
        elif (custom.name == 'Other IHC Markers'):
            patient_dict['other_IHC_markers'] = custom.value
        elif (custom.name == 'Diagnosis of infection'):
            patient_dict['diagnosis_of_infection'] = custom.value
        elif (custom.name == 'Surgery Performed'):
            patient_dict['surgery_performed'] = custom.value
        elif (custom.name == 'Radiotherapy Given'):
            patient_dict['radiotherapy_given'] = custom.value
        elif (custom.name == 'Chemotherapy Given'):
            patient_dict['chemotherapy_given'] = custom.value
        elif (custom.name == 'Immunotherapy Given'):
            patient_dict['immunotherapy_given'] = custom.value
        elif (custom.name == 'Other Therapy Details'):
            patient_dict['other_therapy_details'] = custom.value
    
    
               
    #print(patient_dict)
    return patient_dict
```
